### Task.4. In which year highest number of movies are released ? Show with Bar Graph.

##### dtypes

```python
df.dtypes
```

##### to_datetime

```python
df['Date_N'] = pd.to_datetime(df['release_year'])
```

```python
df.head()
```

```python
df.dtypes
```

##### dt.year.value_counts()

```python
movies_per_year = df['release_year'].value_counts().sort_index()
```

```python
max_year = movies_per_year.max()
```

```python
max_count = movies_per_year.max()
```

##### BAR GRAPH

```python
plt.figure(figsize=(12,6))
movies_per_year.plot(kind='bar')
plt.title(f'Number of Movies released per year (Max:{max_year} with {max_count} movies')
plt.xlabel('year')
plt.ylabel('number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
